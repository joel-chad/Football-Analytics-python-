```python
df.drop(range(47, 57), inplace=True)
```

```python
df = df.drop(["Date", "Time","Day","Venue","Result", "Poss","Attendance","Captain","Formation","Referee","Match Report","Notes"], axis=1)
df1 = df1.drop(["Date", "Time","Day","Venue","Result", "Poss","Attendance","Captain","Formation","Referee","Match Report","Notes"], axis=1)
df2 = df2.drop(["Date", "Time","Day","Venue","Result", "Poss","Attendance","Captain","Formation","Referee","Match Report","Notes"], axis=1)
```

```python
df.dropna(inplace=True)
df1.dropna(inplace=True)
df2.dropna(inplace=True)
```

```python
df = df.loc[df.Comp== "Premier League"]
df1 = df1.loc[df1.Comp== "Premier League"]
df2 = df2.loc[df2.Comp== "Premier League"]
```

```python
TEAM_NAME = "Manchester City"

window = 5
gd_color = "dodgerblue"
xgd_color = "gold"

xg_rolling = df["xG"].rolling(window).mean().values[window:]
xga_rolling = df["xGA"].rolling(window).mean().values[window:]

xg_rolling_1 = df1["xG"].rolling(window).mean().values[window:]
xga_rolling_1 = df1["xGA"].rolling(window).mean().values[window:]

xg_rolling_2 = df2["xG"].rolling(window).mean().values[window:]
xga_rolling_2 = df2["xGA"].rolling(window).mean().values[window:]
```

```python
# Pick up any fonts in the current directory.
# If you do end up downloading the fonts to /usr/share/fonts/truetype,
# change this to: fm.findSystemFonts()
font_files = fm.findSystemFonts('.')
# font_files

# Go through and add each to Matplotlib's font cache.
for font_file in font_files:
    fm.fontManager.addfont(font_file)

# Use your new font on all your plots.
plt.rc('font', family='Quicksand')
```

```python
with plt.style.context("dark_background"):
    plt.rcParams['font.family'] = 'Quicksand' ##set global font
    fig, ax = plt.subplots(1, 3, sharey=True, figsize=(25,8))
    


    ax[0].plot(xg_rolling, color=gd_color,  linestyle="-.", marker="o",  mfc=gd_color, mec="white", markersize=8, mew=0.4, zorder=10)  ##goal-difference
    ax[0].plot(xga_rolling, color=xgd_color,  linestyle="-.", marker = "o", mfc=xgd_color, mec="white", markersize=8, mew=0.4, zorder=10) ##expected goals difference
    ax[0].set_ylim(-.5, 3.5)
    ax[0].grid(linestyle="dashed", lw=0.7, alpha=0.1, zorder=1) ## a faint grid
    
    ax[0].fill_between(x=range(len(xg_rolling)), y1=xg_rolling, y2=xga_rolling, where = xg_rolling>xga_rolling, 
                    alpha=0.2, color=gd_color, interpolate=True, zorder=5) ##shade the areas in between
    ax[0].fill_between(x=range(len(xg_rolling)), y1=xg_rolling, y2=xga_rolling, where = xg_rolling<=xga_rolling, 
                    alpha=0.2, color=xgd_color, interpolate=True, zorder=5)
    
    ax[1].plot(xg_rolling_1, color=gd_color,  linestyle="-.", marker="o",  mfc=gd_color, mec="white", markersize=8, mew=0.4, zorder=10)  ##goal-difference
    ax[1].plot(xga_rolling_1, color=xgd_color,  linestyle="-.", marker = "o", mfc=xgd_color, mec="white", markersize=8, mew=0.4, zorder=10) ##expected goals difference
    ax[1].set_ylim(-.5, 3.5)
    ax[1].grid(linestyle="dashed", lw=0.7, alpha=0.1, zorder=1) ## a faint grid
    
    ax[1].fill_between(x=range(len(xg_rolling_1)), y1=xg_rolling_1, y2=xga_rolling_1, where = xg_rolling_1>xga_rolling_1, 
                    alpha=0.2, color=gd_color, interpolate=True, zorder=5) ##shade the areas in between
    ax[1].fill_between(x=range(len(xg_rolling_1)), y1=xg_rolling_1, y2=xga_rolling_1, where = xg_rolling_1<=xga_rolling_1, 
                    alpha=0.2, color=xgd_color, interpolate=True, zorder=5)
    
    ax[2].plot(xg_rolling_2, color=gd_color,  linestyle="-.", marker="o",  mfc=gd_color, mec="white", markersize=8, mew=0.4, zorder=10)  ##goal-difference
    ax[2].plot(xga_rolling_2, color=xgd_color,  linestyle="-.", marker = "o", mfc=xgd_color, mec="white", markersize=8, mew=0.4, zorder=10) ##expected goals difference
    ax[2].set_ylim(-.5, 3.5)
    ax[2].grid(linestyle="dashed", lw=0.7, alpha=0.1, zorder=1) ## a faint grid
    
    ax[2].fill_between(x=range(len(xg_rolling_2)), y1=xg_rolling_2, y2=xga_rolling_2, where = xg_rolling_2>xga_rolling_2, 
                    alpha=0.2, color=gd_color, interpolate=True, zorder=5) ##shade the areas in between
    ax[2].fill_between(x=range(len(xg_rolling_2)), y1=xg_rolling_2, y2=xga_rolling_2, where = xg_rolling_2<=xga_rolling_2, 
                    alpha=0.2, color=xgd_color, interpolate=True, zorder=5)

        

    for spine in ["top", "right", "left", "bottom"]:
        ax[1].spines[spine].set_visible(False) 
        ax[2].spines[spine].set_visible(False)
        ax[0].spines[spine].set_visible(False)

    # ax.set_position([0.08, 0.08, 0.82, 0.78]) ## make space for the title on top of the axes
    
    ## labels, titles and subtitles
    ax[1].set(xlabel=f"Matchday")     
    ax[1].xaxis.label.set(fontsize=12, fontweight='bold')    

    fig.text(x=0.08, y=0.92, s="Manchester City | xG Performance Trend", 
            ha='left', fontsize=24, fontweight='bold', 
            path_effects=[pe.Stroke(linewidth=3, foreground='0.15'),
                       pe.Normal()])   
    
    fig_text(x=0.08, y=0.90, ha='left',
             fontsize=18, fontweight='bold',
             s='Premier League | 2020-21 | <xG> vs <xGA>',
             path_effects=[pe.Stroke(linewidth=3, foreground='0.15'),
                       pe.Normal()],
             highlight_textprops=[{"color": gd_color},
                                  {"color": xgd_color}])
    
fig.savefig("xg-performance-chart", dpi=180) ##save image
```
