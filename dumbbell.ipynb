```python
import pandas as pd
import matplotlib.pyplot as plt

import matplotlib.patheffects as pe
from adjustText import adjust_text
from highlight_text import fig_text
```

```python
df = pd.read_csv("comparison.csv")
df
```

```python
my_range=range(1,len(df.index)+1)
blue = "dodgerblue"
gold = "gold"
```

```python
with plt.style.context("dark_background"):
    plt.rcParams['font.family'] = 'Helvetica' ##set global font
    fig, ax = plt.subplots(figsize=(12, 8))
    plt.hlines(y=my_range, xmin=df['20/21'], xmax=df['21/22'], color=blue, alpha=0.4)
    plt.scatter(df['20/21'], my_range, color=blue, alpha=1, label='20/21')
    plt.scatter(df['21/22'], my_range, color=gold, alpha=0.8 , label='21/22')
    
    ax.grid(linestyle="dashed", lw=0.7, alpha=0.1, zorder=1) ## a faint grid
    for spine in ["top", "right", "left", "bottom"]:
        ax.spines[spine].set_visible(False)  
        ax.set_position([0.08, 0.08, 0.82, 0.78]) ## make space for the title on top of the axes
    
    # Add title and axis names
    plt.yticks(my_range, df['stat'])
    
    plt.tick_params(left = False, bottom = False)
    
    fig_text(s=f'Data from fbref via Statsbomb',
        x=.65, y =.9, fontsize=14,color='#c7d5cc')

    fig.text(.8,.35,f'@joel_mcfc98', fontstyle='italic',fontsize=10,color='#c7d5cc')
    
    fig.text(x=0.08, y=0.92, s="Manchester City | Premier League", 
            ha='left', fontsize=24, fontweight='bold', 
            path_effects=[pe.Stroke(linewidth=3, foreground='0.15'),
                       pe.Normal()])   
    
    fig_text(x=0.08, y=0.90, ha='left',
             fontsize=18, fontweight='bold',
             s='Comparison of Key Metrics between <20/21> vs <21/22>',
             path_effects=[pe.Stroke(linewidth=3, foreground='0.15'),
                       pe.Normal()],
             highlight_textprops=[{"color": blue},
                                  {"color": gold}])
#     plt.xlabel('Value of the Positions')
#     plt.ylabel('Season')
    
fig.savefig("metric-performance-chart", dpi=180) ##save image
```
